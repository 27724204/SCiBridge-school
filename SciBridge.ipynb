### The problems in observations of the data

The data received had values in column plasma (9999.9) and density (99.99999)   that where either corrupted when the instrument wasn't operating or sensors where over heated by the sun. Furthermore, the column days and years have to be combined to form a correct time of measurement.

```python
#To solve the first problems of correct time of measurement
#First we convert days into years is done by using the formula y = (number of days) / 365
voyage1_data['days_to_years'] = voyage1_data['DOY']/365.242199
voyage2_data['days_to_years'] = voyage2_data['DOY']/365.242199

#Converting to from hours to days
voyage1_data['hours_to_years'] = voyage1_data['Hour']/(24*365.242199)
voyage2_data['hours_to_years'] = voyage2_data['Hour']/(24*365.242199)

#voyage1_data #checking the changes in the data set
voyage1_data
```

```python
#To get the total time all we do is add the time using formula time = year + days to years
voyage1_data['observation_T'] = voyage1_data['Year'] + voyage1_data['days_to_years'] + voyage1_data['hours_to_years']
voyage2_data['observation_T'] = voyage2_data['Year'] + voyage2_data['days_to_years'] + voyage2_data['hours_to_years']

#voyage2_data #checking the changes in the data set 
```

```python
# From since the other columns aren't useful anymore they can be deleted using 
# Drop all columns at once using a list
voyage1_data.drop(columns=['Year', 'DOY', 'Hour', 'hours_to_years', 'days_to_years'], inplace=True)

# Drop all columns at once using a list
voyage2_data.drop(columns=['Year', 'DOY', 'Hour', 'hours_to_years', 'days_to_years'], inplace=True)
```

To clean the errors in plasm column 

```python
import numpy as np # Famous python library used for varies activity
```

```python
#first we replace 9999.9 value with 'nan' meaning nothing
voyage1_data= voyage1_data.replace([9999.9, 99.99999,9999999., 999.9989999999999,999.999, 9999999.0 ], np.nan) 
voyage2_data= voyage2_data.replace([9999.9, 99.99999,9999999., 999.9989999999999,999.999, 9999999.0 ], np.nan) 

r_initial=len(voyage1_data['observation_T'])
r_initial2=len(voyage2_data['observation_T'])

```

```python
#Because the data set is huge, we can't inspect it by eye, but we can use nan to check how many rows have nothing in the data
voyage1_data.isnull().sum() #this is to check those values that are null as stated above
```

```python
#To delete to 'nan' and their rows we use and produces clean set 
voyage1_cleaned_data = voyage1_data.dropna()
voyage2_cleaned_data = voyage2_data.dropna()

r_final=len(voyage1_cleaned_data['observation_T'])
r_final2=len(voyage2_cleaned_data['observation_T'])

voyage1_cleaned_data
```

```python
Loss  = r_initial - r_final
Loss2  = r_initial2 - r_final2

print (f'the total number of Voyager 1 data lost due cleaning: {Loss} ')
print (f'the total number of Voyager 2 data lost due cleaning: {Loss2} ')
```

```python
def Vr_r(r,r0,ro,Vo):
    return Vo*(1-10**((40/3)*((r0 - r)/ro)))
    
#Vr = Vr_r(r,r0,ro,Vo)
```

```python
r0 = 0.005
ro = 1
Vo = 400
r = np.linspace(0,100, 1000000)

Vr = Vr_r(r,r0,ro,Vo)
```

```python
plt.plot(r,Vr,linestyle='--',label = ' model for V$_{o}$ = 400 km/s')
plt.plot(r,Vr/1.6,linestyle='--',label = ' model for V$_{o}$ = 250 km/s')
plt.plot(r,Vr*2,linestyle=':',label = ' model for V$_{o}$ = 800 km/s')

plt.xscale('log')

#plt.xlim(5*10**-3, 8*10**1)
plt.fill_between(r,Vr/1.6, Vr, facecolor = 'lightgrey')
plt.fill_between(r,Vr, Vr*2, facecolor = 'linen')
#plt.text(1*10**-1, 300, r"slow-wind area",color = "black", fontsize= 18)
#plt.text(1*10**-1, 500, r"fast-wind area",color = "black", fontsize= 18)


plt.grid()




#plt.savefig('daily_plasma_speed_data.png',dpi=200,bbox_inches='tight')
```
